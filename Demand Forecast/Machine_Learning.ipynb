```python
import oracledb
connection=oracledb.connect(user="ebizai",password="ebizai",dsn="10.0.2.163:1575/orcl")
async def insert_token(chat_id,user_request,process_name,input_tokens,output_tokens,username):
        cursor=connection.cursor()
        insert_track = """
            INSERT INTO EBIZAI.MST_TRACK_TOKENS (
                VC_CHAT_ID, DT_DATETIME, VC_USER_REQUEST, VC_PROCESS_NAME,
                NU_INPUT_TOKENS, NU_OUTPUT_TOKENS, NU_TOTAL_TOKENS,
                NU_TOTAL_COST, NU_INPUT_TOKEN_COST, NU_OUTPUT_TOKEN_CODE, VC_USER_NAME
            )
            VALUES (
                :chat_id, SYSTIMESTAMP, :user_request, :process_name,
                :input_tokens, :output_tokens, :total_tokens,
                :total_cost, :input_token_cost, :output_token_cost, :user_name
            )
        """

        total_tokens=input_tokens + output_tokens
        INPUT_COST_PER_MILLION = 0.10    
        OUTPUT_COST_PER_MILLION = 0.40  
        input_tokens_cost= ((input_tokens / 1_000_000) * INPUT_COST_PER_MILLION) 
        output_tokens_cost= ((output_tokens / 1_000_000) * OUTPUT_COST_PER_MILLION)
        cost = input_tokens_cost + output_tokens_cost
        
        try:
            cursor.execute(insert_track, {
                "chat_id": chat_id,
                "user_request": user_request,
                "process_name": process_name,
                "input_tokens": input_tokens,
                "output_tokens": output_tokens,
                "total_tokens": total_tokens,
                "total_cost": cost,
                "input_token_cost": input_tokens_cost,
                "output_token_cost": output_tokens_cost,
                "user_name": username
            })
            connection.commit()
        except Exception as e:
              print(e)
        
```
